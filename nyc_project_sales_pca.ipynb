```python
#scaling the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical = ['BOROUGH', 'NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'BLOCK', 'LOT',
       'RESIDENTIAL UNITS', 'COMMERCIAL UNITS', 'TOTAL UNITS',
       'LAND SQUARE FEET', 'GROSS SQUARE FEET', 'YEAR BUILT',
       'TAX CLASS AT TIME OF SALE', 'SALE PRICE']
#features_raw[numerical] = scaler.fit_transform(features_raw[numerical])
data[numerical] = scaler.fit_transform(data[numerical])
```

```python
PCA_dataset_Y= data['SALE PRICE']
```

```python
PCA_dataset_X= data[data.columns.difference(['SALE PRICE'])]
```

```python
PCA_dataset_X
```

```python
from sklearn.decomposition import PCA
# TODO: Apply PCA by fitting the good data with the same number of dimensions as features
pca = PCA(n_components=3)
pca.fit(PCA_dataset_X)

# TODO: Transform log_samples using the PCA fit above
#pca_samples = pca.transform(log_samples)

# Generate PCA results plot
#pca_results = vs.pca_results(good_data, pca)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=7)
principalComponents = pca.fit_transform(PCA_dataset_X)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2','PC3', 'PC4', 'PC5', 'PC6','PC7'])
```

```python
finalDf = pd.concat([principalDf, PCA_dataset_Y], axis = 1)
```

```python
#from he graph we can say that pca stops at 5 because after thatit is significantly low
percent_variance = np.round(pca.explained_variance_ratio_* 100, decimals =2)
columns = ['PC1', 'PC2','PC3', 'PC4','PC5', 'PC6', 'PC7']
plt.bar(x= range(1,8), height=percent_variance, tick_label=columns)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Screen Plot')
plt.show()
```

```python
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(PCA_dataset_X)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2','PC3', 'PC4'])
```

```python
plt.scatter(finalDf.PC1, finalDf.PC2)
plt.title('PC1 against PC2')
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
```

```python
#outlier analysis using zscore
from scipy import stats
z = np.abs(stats.zscore(finalDf))
df = finalDf[(z < 3).all(axis=1)]
df.head()
```
